## 3) Discretización

    ```
    data['ingreso_total_lab'] = pd.qcut(data['ingreso_total_lab'], q=10, duplicates='drop')

    ```

- Para la columna `edad` discreticen usando igual distancia con `bins=5`.

````python
df.copy
````

````python
# Discretizamos: ingresos_familiares
ingresos_familiares_cat=pd.qcut(df['ingresos_familiares'], q=8)
````

````python
df['ingresos_familiares']
````

````python
print(ingresos_familiares_cat.value_counts())
status(ingresos_familiares_cat)
````

````python
# Discretizamos: ingreso_per_capita_familiar

ingreso_per_capita_familiar_cat = pd.qcut(df['ingreso_per_capita_familiar'], q=10)
````

````python
df['ingreso_per_capita_familiar']
````

````python
print(ingreso_per_capita_familiar_cat.value_counts())
status(ingreso_per_capita_familiar_cat)
````

````python
#Para las siguientes columnas ingreso_total_lab y ingreso_total_no_lab consideren:

df['ingreso_total_lab'] = pd.qcut(df['ingreso_total_lab'], q=10, duplicates='drop')
df['ingreso_total_no_lab'] = pd.qcut(df['ingreso_total_no_lab'], q=4, duplicates='drop')
````

````python
# Para la columna edad discreticen usando igual distancia con bins=5.

edad_cat=pd.cut(df['edad'], bins=5)
````

````python
edad_cat.value_counts()
````

````python
# CREO UN NUEVO DATA SET Y AGREGO LAS COLUMNAS NUEVAS
df2 = df
df2['ingreso_per_capita_familiar'] = ingreso_per_capita_familiar_cat
df2['ingresos_familiares'] = ingresos_familiares_cat
df2['edad'] = edad_cat
````

````python
status(df2)
````

## 4) Preparación de datos

- Cambien el tipo de dato a `str` de las siguientes columnas: `comuna` y `nhogar`.
  <br>_¿Por qué hacemos esto?_ Para estas columnas el número es simplemente una connotación, para representar una comuna por ejemplo pero no hay una relación numérica entre ellos.

````python
df2['comuna'] = df2['comuna'].astype(str)
````

````python
df2['nhogar'] = df2['nhogar'].astype(str)
````

- _¿Qué esperas como valor en la columna `años_escolaridad`?_ Número enteros pero no siempre es así, cada entidad o empresa tiene diferentes formas de rellenar una encuesta.
    <br>Evalua lo siguiente: `data['años_escolaridad'].unique()`, vas a poder ver los valores únicos en la columna. Donde destacamos que todos son `object/string`.


````python
df2['años_escolaridad'].unique()
````

- Reemplazar `Ningun año de escolaridad aprobado` por un '0'.

````python
#Si hacemos esto de una, marca error:
df2['años_escolaridad'] = df2['años_escolaridad'].replace('Ningun año de escolaridad aprobado', '0')
#Primero lo pasamos a entero y después hacemos el replace.
df2['años_escolaridad'] = df2['años_escolaridad'].astype(float).astype("Int32")
````

````python
print(f'Los datos nulos son: {df2['años_escolaridad'].isna().sum()}')
df2['años_escolaridad'].unique()
````

- Discreticen para la columna `años_escolaridad` por igual frecuencia e igual rango con un `q=5`

````python
años_escolaridad_cat = pd.qcut(df2['años_escolaridad'], q=5)
````

````python
años_escolaridad_cat.value_counts()
````

````python
# AGREGO LAS COLUMNAS NUEVAS Y ELIMINO LAS ANTERIORES

df['años_escolaridad'] = años_escolaridad_cat
status(df2)
````

- No necesariamente siempre hay que rellenar los `NaN` en todas las columnas, porque quizás esa cantidad de `NaN` no es tan representativa para nuestro análisis. Así que podes eliminarlo para todo el dataframe o para ciertas columnas.

````python
# Eliminar filas que contengan NaN
df2 = df2.dropna(subset=['situacion_conyugal', 'sector_educativo', 'lugar_nacimiento', 'afiliacion_salud'])
````

- Después de eliminar filas, podes resetear el índice para que mantenga la secuencia:

````python
df2 = df2.reset_index(drop=True)
````

- Rellenar los datos faltantes para la columna `años_escolaridad`. Primero añadan la categoría `desconocido` y luego hacen un rellenado de los datos faltantes con `desconocido`.

````python
# Primero: agregar la nueva categoría al dtype categórico
df2['años_escolaridad'] = df2['años_escolaridad'].cat.add_categories('desconocido')

# Segundo: ahora sí rellenar los NaN
df2['años_escolaridad'] = df2['años_escolaridad'].fillna('desconocido')
````


- Rellenar los datos faltantes para la columna `nivel_max_educativo` con `value=desconocido`

````python
df2['nivel_max_educativo'].unique()
````

````python
df2['nivel_max_educativo'] = df2['nivel_max_educativo'].fillna('desconocido')
````
